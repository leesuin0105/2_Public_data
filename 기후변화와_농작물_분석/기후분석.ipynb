# 0. 전처리

```python
# 데이터프레임 리스트 생성
dataframes = [기후2011, 기후2012, 기후2013, 기후2014, 기후2015, 기후2016, 기후2017, 기후2018, 기후2019, 기후2020, 기후2021]

# 데이터프레임을 관측지점별로 합침
총_기후 = pd.concat(dataframes)

# '합계수평면일사 (MJ/㎡)' 열 삭제
총_기후 = 총_기후.drop('합계수평면일사 (MJ/㎡)', axis=1)
```

```python
# 중간 결과 저장
기후 = 총_기후.copy()
```

```python
# 시점 열을 float64에서 문자열로 변환합니다.
기후['시점'] = 기후['시점'].astype(str)

# 시점 열의 문자열을 연도와 월로 분리합니다.
기후['연도'] = 기후['시점'].str.split('.').str[0]
기후['월'] = 기후['시점'].str.split('.').str[1]

# '월' 열의 값을 1인 경우에는 10으로 바꿉니다.
기후['월'] = 기후['월'].apply(lambda x: '10' if x == '1' else x)

# 연도, 월, 일을 합쳐서 새로운 '날짜' 열을 생성합니다.
기후['날짜'] = 기후['연도'] + '-' + 기후['월'] + '-01'

# '날짜' 열을 datetime 형식으로 변환합니다.
기후['날짜'] = pd.to_datetime(기후['날짜'], format='%Y-%m-%d')

# '연도', '월' 열 및 이전 '시점' 열을 삭제합니다.
기후.drop(['연도', '월', '시점'], axis=1, inplace=True)

# 이제 '날짜' 열이 datetime 형식으로 변경됨
```

```python
# 숫자 열의 데이터 타입을 적절한 숫자 형식으로 변환합니다.
numeric_columns = ['평균기온 (℃)', '평균최고기온 (℃)', '평균최저기온 (℃)', '최고기온 (℃)', '최저기온 (℃)', '합계강수량 (mm)', '평균풍속 (m/s)', '평균상대습도 (%)', '합계일조시간 (hr)', '평균전운량 (할)', '평균지면온도 (℃)']
기후[numeric_columns] = 기후[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
결측값_개수 = 기후.isna().sum()
결측값_개수
```

```python
기후.drop('연도', axis=1, inplace=True)
```
